### Objective of the problem.


<p>The dataset contains cases from a study that was conducted between 
1958 and 1970 at the <br>University of Chicago's Billings Hospital on
the survival of patients who had undergone surgery for breast
cancer.

Objective of the problem is to find whether a new pateint who has undergone surgery for breast cancer <br>
will survive 5 years or longer<br>
or<br>
die within 5 years after the surgery.

It is basically a supervised bi-classification problem hence we have two labels in our taget variable (Survival status) i.e., <br>
1 = the patient survived 5 years or longer <br>
2 = the patient died within 5 year

For the purpose of classifiaction we make use of three features namely :<br>
age	(Age of patient at time of operation (numerical))<br>
operation_Year	(Patients year of operation (year - 1900, numerical))<br>
axil_nodes	(Number of positive axillary nodes detected (numerical)) <p>

### (b) CDF AND PDF

```python
import numpy as np
ss_1 = df.loc[df["survival_status"] == 1];
ss_2 = df.loc[df["survival_status"] == 2];

counts, bin_edges = np.histogram(ss_1['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.title('PDF and CDF of feature - Age')
plt.xlabel("age")
plt.ylabel("Probability of RV X<=x for CDF, Probability of RV X<=x for PDF ")


legend_drawn_flag = True
plt.legend(["PDF Plot", "CDF Plot"], loc=0, frameon=legend_drawn_flag)




plt.show();
```

```python
counts, bin_edges = np.histogram(ss_1['axil_nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


plt.title('PDF and CDF of feature axil_nodes')
plt.xlabel("axil_nodes count")
plt.ylabel("Probability of RV X<=x for CDF, Probability of RV X<=x for PDF ")

location = 0 
legend_drawn_flag = True
plt.legend(["PDF Plot", "CDF Plot"], loc=0, frameon=legend_drawn_flag)

plt.show();
```

```python
counts, bin_edges = np.histogram(ss_1['operation_year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


plt.title('PDF and CDF of feature - Operation_year')
plt.xlabel("operation_year")
plt.ylabel("Probability of RV X<=x for CDF, Probability of RV X<=x for PDF ")

location = 0 
legend_drawn_flag = True
plt.legend(["PDF Plot", "CDF Plot"], loc=0, frameon=legend_drawn_flag)


plt.show();
```

2.All the plots are drawn by taking survival status as 1 into consideration.
    
3.PDF represent how many points exists in a range wihle CDF rrepresents percentage of points that are less than of equal to a certain value.
    
4.When plotted against age, 16% of people are aged between 53.5 and 58.2
